You can also run a whole cell with timeit using double-percent `%%timeit`

```
N = 100  # number of points in my plot
```

```
#%%timeit    # try uncommenting
X, Y = [], []
for i in range(N+1):
    val = 2 * math.pi * i/N
    X.append(val)
    Y.append(math.sin(val))
```

```
#%%timeit    
X2 = [2.0 * math.pi * i/N for i in range(N + 1)]
Y2 = [math.sin(val) for val in X2]
```

```
#%%timeit
X3 = 2.0 * np.pi * np.arange(N + 1)/N
Y3 = np.sin(X3)
```

```
#%%timeit
X4 = np.linspace(0, 2.0 * np.pi, N + 1)
Y4 = np.sin(X4)
```

```
#%%timeit
dx = 2*np.pi/N
X5 = np.arange(0, 2.0 * np.pi + dx, dx)
Y5 = np.sin(X5)
```

```
plt.scatter(X, Y,   s=50, label="method 1")
plt.scatter(X2, Y2, s=40, label="method 2")
plt.scatter(X3, Y3, s=30, label="method 3")
plt.scatter(X4, Y4, s=20, label="method 4")
plt.scatter(X5, Y5, s=10, label="method 5")
plt.legend()
```

Try increasing N...

* Results are not always obvious! 
* When in doubt, go for the easiest to read if the performance is "fine"
